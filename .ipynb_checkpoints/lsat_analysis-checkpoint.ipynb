```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import glob

#change default figure size
from pylab import rcParams
rcParams['figure.figsize'] = 12, 12

score_files = glob.glob('*.txt')
Q,S,scores=[],[],[]

# read question types
question_types = {}
with open('index/question_type.txt', 'r') as f:
    for line in f.readlines():
        l = line.strip().split(',')
        question_types[l[0]] = l[1]

# read score files
for i, score_file in enumerate(score_files):        
    Q.append({})
    S.append({})
    
    with open(score_file, 'r') as f:
        for line in f.readlines():
            l=line.strip()
            if not l: continue

            sp = l.split(',')    
            # score is the header
            if len(sp) == 1:
                scores.append(int(sp[0]))
            else:
                if sp[1] not in Q[i]:
                    Q[i][sp[1]] = 0
                    S[i][sp[1]] = 0

                Q[i][sp[1]] += 1
                S[i][sp[1]] += int(sp[2])
```

```python
y = []
x1 = []
x2 = []

for key in Q:
    y.append(question_types[key])
    x1.append(Q[key])
    x2.append(S[key])
    
ind = np.arange(len(y))
width = 0.35       # the width of the bars    

# fig = plt.figure()
fig, ax = plt.subplots(2,1)

# r1 = ax.barh(ind, y1, width, color='green')
# r2 = ax.barh(ind+width, y2, width, color='yellow')

# ax.set_ylabel('Question types')
# ax.set_xlabel('Number of questions')
# ax.set_yticks(ind+width)
# ax.set_yticklabels(y)
# ax.legend( (r1[0], r2[0]), ('Total', 'Correct') )

# plt.show()

r1 = ax[0].barh(ind, y1, width, color='green')
r2 = ax[0].barh(ind+width, y2, width, color='yellow')

ax[0].set_title('41.txt')
ax[0].set_ylabel('Question types')
ax[0].set_xlabel('Number of questions')
ax[0].set_yticks(ind+width)
ax[0].set_yticklabels(y)
ax[0].legend( (r1[0], r2[0]), ('Total', 'Correct') )

plt.show()
```
